```python

from wordcloud import WordCloud
import PIL.Image as image
import numpy as np
import jieba

with open("wordcloud.txt",encoding="utf-8") as fp:
    text = fp.read()
    
    word_list = jieba.cut(text)
    
    text = " ".join(word_list)
   
    # mask = np.array(image.open("wordcloud.jpg"))
    wordcloud = WordCloud( font_path="C:\Windows\Fonts\SIMHEI.ttf", # Windows上运行需要中文字体库
                            background_color='white'    ).generate(text)
    image_produce = wordcloud.to_image()
    image_produce.show()
    wordcloud.to_file('word_cloud.jpg') # 保存词云图片
```
