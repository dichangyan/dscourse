### 6、3d图

```python

import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

# 创建3d图形的两种方式
# ax = Axes3D(fig)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(111, projection='3d')

# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    # x-y 平面的网格
R = np.sqrt(X ** 2 + Y ** 2)

# height value
Z = np.sin(R)

# rstride:行之间的跨度  cstride:列之间的跨度
# rcount:设置间隔个数，默认50个，ccount:列的间隔个数  不能与上面两个参数同时出现
#vmax和vmin  颜色的最大值和最小值

ax0=ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap( 'rainbow'))
# zdir : 'z' | 'x' | 'y' 表示把等高线图投射到哪个面
# offset : 表示等高线图投射到指定页面的某个刻度

'''
matplotlib中的色系
cmaps['Sequential'] = [
            'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']

'''


ax2=ax.contourf(X,Y,Z,zdir='z',offset=-2)
# 设置图像z轴的显示范围，x、y轴设置方式相同
ax.set_zlim(-2,2)
plt.colorbar(ax0)
plt.colorbar(ax2)
plt.show()
```

### 补充：关于args and kwargs**
> 1. args类型是一个tuple，而kwargs则是一个字典dict，并且args只能位于kwargs的前面。
> 2. args和kwargs组合起来可以传入任意的参数，这在参数未知的情况下是很有效的，同时加强了函数的可拓展性。

```python
#*args 中的*在函数定义时表示pack，将多个参数打包成一个tuple
def foo(a, *args):
    print('a:', a)
    print('args:', args)
    print(type(args))

foo(1, 2, 3, 4, 5)
```

```python
def foo(a, b, *number):
    print('a:', a)
    print('b:', b)
  
    print('number:', number)
    for i in number:
        print(i)
    print(type(number))


foo(1, 2, 3, 4, 5)
```

```python
#*args 中的*在函数调用时表示unpack，将一个数组，拆包
def bar(a,b,c):
    print(a,b,c)

bar(*[1,2,3])
```

```python
#*args 中的*在函数定义时表示pack，将多个带有关键字的参数，打包成一个字典
def test_kwargs(first, *args, **kwargs):
   print('Required argument: ', first)
   print(type(kwargs))
   for v in args:
      print ('Optional argument (args): ', v)
   for k, v in kwargs.items():
      print ('Optional argument %s (kwargs): %s' % (k, v))

test_kwargs(1, 2, 3, 4, k1=5, k2=6)
```

```python
def test_args_kwargs(arg1, arg2, arg3):
    print("arg1:", arg1)
    print("arg2:", arg2)
    print("arg3:", arg3)

args = ("two", 3, 5)

test_args_kwargs(*args)

kwargs = {"arg3": 3, "arg2": "two", "arg1": 5}

test_args_kwargs(**kwargs)  #*args 中的*在函数调用时表示unpack，将一个字典,拆包，对应多个参数
```
